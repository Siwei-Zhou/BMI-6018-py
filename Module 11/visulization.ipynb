```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/sample_data/diabetic_data.csv')
```

1

```python
# Readmission status of different admission type
# x - age range
# y - medication number
sns.color_palette("mako", as_cmap=True)
sns.relplot(
    data=data,
    x="age", y="num_medications", col="admission_type_id",
    hue="readmitted", style="readmitted"
)
```

```python
#The porpotion of race.
plt.figure(figsize=(8,8))
y = np.array([76099, 19210, 2273, 2037, 1506, 641])
mylabels = ["Caucasian", "AfricanAmerican", "?", "Hispanic", "Other", "Asian"]
mycolors = ["#83cbac", "#1ba784", "#93d5dc", "#ed9db2","#f2ce2b", "#fffefa"]
plt.pie(y, labels = mylabels,colors=mycolors,
        textprops={'fontsize': 10, 'color': 'k'},
        wedgeprops=dict(width=0.4, edgecolor='w'),
        autopct='%.2f%%')
plt.legend(title = "Races:")
plt.title("The porpotion of race")
plt.show() 
```
