```python
# -*- coding: utf-8 -*-

#%% Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#%% Importing Data
flights_data = pd.read_csv('/content/sample_data/flights.csv')
flights_data.head(10)
weather_data_pd = pd.read_csv('/content/sample_data/weather.csv')
weather_data_np = weather_data_pd.to_numpy()
```

```python
#%% Pandas Data Filtering/Sorting Question Answering
#use flights_datav
#Question 1 How many flights were there from JFK to SLC? 
q_1 = flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')]
print(q_1.shape[0])

#Question 2 How many airlines fly to SLC? Should be int
q_2 = flights_data[flights_data['dest'] == 'SLC']
print(q_2.shape[0])

#Question 3 What is the average arrival delay for flights to RDU? float
q_3 = flights_data[flights_data['dest'] == 'RDU']
print(q_3['arr_delay'].mean())

#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
SEA = flights_data[flights_data['dest'] == 'SEA'].shape[0]
LGA = flights_data[(flights_data['origin'] == 'LGA') & (flights_data['dest'] == 'SEA')].shape[0]
JFK = flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SEA')].shape[0]
q_4_LGA = LGA / SEA
q_4_JFK = JFK / SEA
print('LGA:', q_4_LGA, '  JFK:', q_4_JFK)

#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)
flights_data['date'] = flights_data['year'].map(str)+"/"+flights_data['month'].map(str)+"/"+flights_data['day'].map(str)
pd.to_datetime(flights_data['date'], format = "%Y/%m/%d")
flights_data.groupby(by='date',axis=0)
flights_data.groupby(by='date',axis=0).mean()
s = flights_data.groupby(by='date',axis=0)['dep_delay'].mean()
flights_data['dep_delay_mean'] = flights_data['date'].map(s.to_dict())
i = pd.DataFrame(flights_data,columns=['date','dep_delay_mean'])
q_5=i.drop_duplicates(subset=None, keep='first', inplace=False)
print(q_5[(q_5['dep_delay_mean'] == q_5['dep_delay_mean'].max())])

#Question 6 Which date has the largest average arrival delay? pd slice with date and floatq_6 
flights_data = pd.read_csv('/content/sample_data/flights.csv')
flights_data['date'] = flights_data['year'].map(str)+"/"+flights_data['month'].map(str)+"/"+flights_data['day'].map(str)
pd.to_datetime(flights_data['date'], format = "%Y/%m/%d")
flights_data.groupby(by='date',axis=0)
t = flights_data.groupby(by='date',axis=0)['arr_delay'].mean()
flights_data['arr_delay_mean'] = flights_data['date'].map(t.to_dict())
j = pd.DataFrame(flights_data,columns=['date','arr_delay_mean'])
q_6=j.drop_duplicates(subset=None, keep='first', inplace=False)
print(q_6[(q_6['arr_delay_mean'] == q_6['arr_delay_mean'].max())])


#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
flights_data['speed'] = flights_data['distance'] / flights_data['air_time']
q_7 = flights_data.loc[((flights_data['origin'] == 'LGA') | (flights_data['origin'] == 'JFK')) & (flights_data['year'] == 2013), ["tailnum", "speed"]]
print(q_7[(q_7['speed'] == q_7['speed'].max())])

#Question 8 Replace all nans in the weather pd dataframe with 0s. Pd with no nans
q_8 = weather_data_pd.fillna(0)
print(q_8)
```
